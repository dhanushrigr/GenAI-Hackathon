```
import pandas as pd
import numpy as np

# Simulate a dataset
np.random.seed(42)
users = [f"User_{i}" for i in range(1, 101)]
products = [f"Product_{i}" for i in range(1, 21)]

data = {
    "user_id": np.random.choice(users, 1000),
    "product_id": np.random.choice(products, 1000),
    "rating": np.random.randint(1, 6, size=1000),
    "timestamp": pd.date_range(start="2023-01-01", periods=1000, freq="min")
}

df = pd.DataFrame(data)
df.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encode users and products
user_encoder = LabelEncoder()
product_encoder = LabelEncoder()

df['user_id'] = user_encoder.fit_transform(df['user_id'])
df['product_id'] = product_encoder.fit_transform(df['product_id'])

# Split data
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dot, Flatten, Dense

# Define model
n_users = len(user_encoder.classes_)
n_products = len(product_encoder.classes_)
embedding_size = 50

# User embedding
user_input = Input(shape=(1,))
user_embedding = Embedding(n_users, embedding_size)(user_input)
user_vec = Flatten()(user_embedding)

# Product embedding
product_input = Input(shape=(1,))
product_embedding = Embedding(n_products, embedding_size)(product_input)
product_vec = Flatten()(product_embedding)

# Dot product for interaction
dot_product = Dot(axes=1)([user_vec, product_vec])

# Model
model = Model(inputs=[user_input, product_input], outputs=dot_product)
model.compile(optimizer='adam', loss='mse')
model.summary()

# Training
train_x = [train['user_id'], train['product_id']]
train_y = train['rating']

test_x = [test['user_id'], test['product_id']]
test_y = test['rating']

model.fit(train_x, train_y, epochs=10, batch_size=64, validation_data=(test_x, test_y))
```

```
# Function to recommend products for a given user
def recommend_products(user_id, top_n=5):
    user_idx = user_encoder.transform([user_id])[0]
    product_indices = np.arange(n_products)
    predictions = model.predict([np.array([user_idx] * n_products), product_indices])
    product_ratings = pd.DataFrame({
        "product_id": product_indices,
        "predicted_rating": predictions.flatten()
    })
    top_products = product_ratings.sort_values("predicted_rating", ascending=False).head(top_n)
    return product_encoder.inverse_transform(top_products['product_id'])

# Example recommendation
recommend_products("User_5", top_n=5)
```

```
# Simulate feedback
feedback = {
    "user_id": user_encoder.transform(["User_5"])[0],
    "product_id": product_encoder.transform(["Product_2"])[0],
    "rating": 5
}
feedback_df = pd.DataFrame([feedback])

# Retrain model
new_data = pd.concat([train, feedback_df])
train_x = [new_data['user_id'], new_data['product_id']]
train_y = new_data['rating']
model.fit(train_x, train_y, epochs=2, batch_size=32)
```
